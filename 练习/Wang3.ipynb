```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.title('学生学习能力对比')
x = np.arange(1,11,1)
plt.plot(x,x*1.2)
plt.plot(x,x*2)
plt.plot(x,x*3)
plt.plot(x,x*4)
plt.legend(['黄亚兰','周兰','胡飞','赵云'])
plt.show()
```

```python
GDP = [28000 , 30133 , 18590 , 19500]
plt.bar(range(4),GDP,align = 'center', color = 'blue',alpha = 1)
plt.ylabel('GDP') #添加轴坐标
plt.title('2021年四个直辖市的GDP对比') #标题
plt.xticks(range(4),['北京市', '上海市', '天津市', '重庆市'],rotation = 30) # 添加刻度标签
plt.ylim([5000, 35000]) # plt.ylim([bottom, top])设置y轴范围值，bottom下限值，top上限值
plt.show()
```
